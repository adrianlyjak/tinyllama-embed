```python
import requests
import json

# Define a list of AI-related terms to broaden the search
ai_related_terms = [
    "Artificial Intelligence",
    "Machine Learning",
    "Deep Learning",
    "Neural Networks",
    "Natural Language Processing",
    "Transformer",
    "Embeddings",
]

query = " | ".join([f'"{term}"' for term in ai_related_terms])

response = requests.get("https://api.semanticscholar.org/graph/v1/paper/search/bulk", params={
  "query": query,
  "s2FieldsOfStudy": "Computer Science",
  "publicationDateOrYear": "2023-06-01:",
  "fields": "title,publicationVenue,year,authors,externalIds,abstract,referenceCount,citationCount,influentialCitationCount,fieldsOfStudy,s2FieldsOfStudy,publicationTypes,publicationDate,journal",
})
response.raise_for_status()
json_response = response.json()
length = len(json_response["data"])
print(f"length: {length}")
print(json.dumps(json_response, indent=2))
```
